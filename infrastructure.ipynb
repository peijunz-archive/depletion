# Test Segm Classes
## Test Lines

```python
L=Line([0.5, -0.5], [0.5, 0.5])
R=L.reverse()
print("L is {}, R is {}".format(L, R))
assert(norm2(L.end()-L.start()+R.end()-R.start())<1e-10)
assert(repr(eval(repr(L)))==repr(L))
```

```python
for i in range(0, 4):
    L.rotate(i*pi/2).draw()
ll, lr=L.split([0.7, 0])
ll.draw(label="Segm 1")
lr.draw(label="Segm 2")
L.extend(0.2).draw('--', label='Extended')
E=Line([0.53, -0.5], [0.64, 0.5])
E.draw()
annotate('O1', xy=Segm.intersect(ll, E)[0])
annotate('O2', xy=Segm.intersect(lr, E)[0])
axis('equal');
legend();
```

## Circle Class

```python
C=Circle(1, [0, 0], [0, pi/2])
D=C.reverse()
print("C is {}\nD is {}".format(C, D))
assert(norm2(C.end()-C.start()+D.end()-D.start())<1e-10)
assert(repr(eval(repr(C)))==repr(C))
```

```python
for i in range(1, 4):
    C.rotate(i*pi/2).draw()
ll, lr=C.split([3,4])
ll.draw(label="Segm 1")
lr.draw(label="Segm 2")
E=C.extend(0.1)
E.draw(label="Extended")
C.extend(-0.1).draw(label="Shrinked")
S=C.shift([0.2, 0])
S.draw(label="Shift")
LL=Line([0.75, 0.2], [1.5, 0.5])
LL.draw()
annotate('O1', xy=Segm.intersect(LL, S)[0]);
axis('equal');
legend()
annotate('O1', xy=Segm.intersect(S, E)[0]);
```
